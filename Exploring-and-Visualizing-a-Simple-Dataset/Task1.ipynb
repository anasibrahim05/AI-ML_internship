```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv("Iris.csv")
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x='SepalLengthCm',
    y='PetalLengthCm',
    hue='Species',
    s=100
)

plt.title("Sepal Length vs Petal Length")
plt.show()
```

```python
df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].hist(
    figsize=(10,8),
    bins=15
)

plt.suptitle("Feature Distributions")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])

plt.title("Box Plot for Iris Features")
plt.show()
```
